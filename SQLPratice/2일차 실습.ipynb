```python
%%sql

SELECT channel
     , COUNT(1)    cnt                -- 채널별 카운트를 하고 싶은 경우. COUNT 함수!!
 FROM raw_data.user_session_channel
GROUP BY channel
ORDER BY cnt DESC;
```

```python
%%sql

SELECT CASE WHEN channel in ('Facebook', 'Instagram') 
            THEN 'Social-Media'
            WHEN channel in ('Google', 'Naver') 
            THEN 'Search-Engine'
            ELSE 'Something-Else'
             END channel_type
    , COUNT(*) channel_type_cnt        
 FROM raw_data.user_session_channel
GROUP BY channel_type;
```
